```python
rating = pd.read_csv('BX-Book-Ratings.csv', sep=';', error_bad_lines=False, encoding="latin-1")
rating.columns = ['userID', 'ISBN', 'bookRating']
```

```python
rating.head(3)
```

```python
rating.tail()
```

```python
print("Columns pada ratings")
print(rating.columns,'\n')

print("Shape :")
print(rating.shape,'\n')

print("Information:")
print(rating.info(),'\n')

print(rating.isnull().sum())
```

```python
seq = rating.bookRating.value_counts().reset_index()
seq
```

```python
sns.catplot(x="index", y="bookRating", kind="bar", data=seq,height=6, aspect=30/10)
```
